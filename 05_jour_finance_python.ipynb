## 1. Listes Python
Comprendre les opérations de liste de base pour le stockage des données financières.

```python
# Creating a list of stock prices
prices = [105.25, 106.00, 104.50, 107.25, 108.00]
                
# Calculating daily returns
daily_returns = [(prices[i] - prices[i-1])/prices[i-1] * 100 
                 for i in range(1, len(prices))]
```

```python
# Affiche les rendements quotidiens en pourcentage
print("Daily Returns:", daily_returns)
```

## 2. Tableaux NumPy
Utilisation de tableaux NumPy pour des calculs financiers efficaces.

```python
import numpy as np

# Converting stock prices to numpy array
prices_array = np.array([105.25, 106.00, 104.50, 107.25, 108.00])

# Calculating returns using numpy
returns = np.diff(prices_array) / prices_array[:-1] * 100
print("Numpy Daily Returns:", returns)

# Calculate volatility (standard deviation of returns)
volatility = np.std(returns)
print("Volatility :", volatility)
```

## 5. Les formules

**Formules de Moyenne Mobile et Signal**

**1. Moyenne Mobile Simple (SMA)**

La **moyenne mobile simple** sur \(n\) périodes est :

$$
\text{SMA}_n(t) = \frac{1}{n} \sum_{i=0}^{n-1} P_{t-i}
$$

**Définitions :**  
- \(P_t\) : prix ou valeur à la période \(t\)  
- \(n\) : nombre de périodes  
- \(t\) : période actuelle  

**2. Moyenne Mobile Exponentielle (EMA)**

La **moyenne mobile exponentielle** sur \(s\) périodes est :

$$
\text{EMA}_t = \alpha \cdot P_t + (1-\alpha) \cdot \text{EMA}_{t-1}
$$

où :

$$
\alpha = \frac{2}{s+1}
$$

**Définitions :**  
- \(P_t\) : prix ou valeur à la période \(t\)  
- \(s\) : période de l’EMA  
- \(\alpha\) : coefficient de lissage  

**3. MACD**

Le **MACD** est la différence entre deux EMA (rapide et lente) :

$$
\text{MACD}(t) = \text{EMA}_{\text{fast}}(t) - \text{EMA}_{\text{slow}}(t)
$$

**Définitions :**  

- \text{EMA}_{\text{fast}}(t) : EMA sur la période courte (ex. 12)  
- \(\text{EMA}_{\text{slow}}\) : EMA sur la période longue (ex. 26)  


**4. Signal (Ligne de Signal du MACD)**

Le **signal** est l’EMA du MACD sur \(s_{\text{signal}}\) périodes :

$$
\text{Signal}(t) = \text{EMA}_{s_{\text{signal}}}(\text{MACD}(t))
$$

**Définitions :**  
- \(s_{\text{signal}}\) : période de l’EMA du signal (ex. 9)  
- \(\text{MACD}(t)\) : valeur du MACD à la période \(t\)  

**5. Histogramme MACD**

L’histogramme représente la différence entre le MACD et sa ligne de signal :

$$
\text{Histogram}(t) = \text{MACD}(t) - \text{Signal}(t)
$$

**Définitions :**  
- Indique la force et la direction du momentum.  
- Positive : MACD au-dessus du signal (momentum haussier).  
- Négative : MACD en dessous du signal (momentum baissier).

```python
#2. Mettez en place un calculateur de moyenne mobile simple :
#- Créer une liste de prix historiques
#- Calculer les moyennes mobiles sur 5 jours et 20 jours
#- Générer des signaux lorsque la MA courte croise la MA longue


import matplotlib.pyplot as plt

# Jeu de données (monte puis descend pour forcer Buy & Sell)
prices = [
    100, 99, 98, 97, 96,
    98, 100, 103, 106, 110,  # montée => Buy
    112, 115, 117, 118, 119,
    117, 114, 111, 108, 105,  # descente => Sell
    103, 101, 99, 98, 97
]

def moving_average(prices, window):
    return [sum(prices[i:i+window]) / window for i in range(len(prices) - window + 1)]

# Calcul des MA
ma5 = moving_average(prices, 5)
ma20 = moving_average(prices, 20)
print("5-Day MA:", ma5)
print("20-Day MA:", ma20)

# Indices pour aligner
ma5_x = list(range(4, len(ma5) + 4))
ma20_x = list(range(19, len(ma20) + 19))

# Alignement
ma5_aligned = ma5[15:]
ma5_x_aligned = ma5_x[15:]

# Détection des signaux
signals = []
for i in range(1, len(ma20)):
    if ma5_aligned[i] > ma20[i] and ma5_aligned[i-1] <= ma20[i-1]:
        signals.append((ma20_x[i], 'Buy'))
    elif ma5_aligned[i] < ma20[i] and ma5_aligned[i-1] >= ma20[i-1]:
        signals.append((ma20_x[i], 'Sell'))

print("Signals:", signals)

# --- Graphe vertical ---
plt.figure(figsize=(20, 5))  # largeur petite, hauteur très grande
plt.plot(prices, label='Prices', linewidth=2)
plt.plot(ma5_x, ma5, label='5-Day MA')
plt.plot(ma20_x, ma20, label='20-Day MA')

# Signaux
for sig_day, sig_type in signals:
    plt.scatter(sig_day, prices[sig_day],
                color='green' if sig_type == 'Buy' else 'red',
                marker='^' if sig_type == 'Buy' else 'v',
                s=120, label=sig_type if sig_day == signals[0][0] else "")
    plt.annotate(sig_type,
                 (sig_day, prices[sig_day]),
                 textcoords="offset points",
                 xytext=(0, 15),
                 ha='center',
                 color='green' if sig_type == 'Buy' else 'red',
                 fontsize=10,
                 fontweight='bold')

plt.legend()
plt.title("Moyennes Mobiles (5 & 20 jours) avec Signaux Buy & Sell")
plt.show()
```

```python
# 3. Créez un calculateur d’obligations :
#- Stockez les détails de l’obligation (valeur nominale, coupon, échéance) dans des listes imbriquées
#- Calculer le rendement à l’échéance
#- Comparer plusieurs liaisons à l’aide d’opérations de liste
# Exemple de solution ci-dessus:
import numpy as np

# 1. Stockage des obligations
bonds = [
    {'name': 'Bond A', 'face_value': 1000, 'coupon_rate': 0.05, 'years_to_maturity': 10},
    {'name': 'Bond B', 'face_value': 1000, 'coupon_rate': 0.04, 'years_to_maturity': 5},
    {'name': 'Bond C', 'face_value': 1000, 'coupon_rate': 0.06, 'years_to_maturity': 15}
]

# Prix du marché actuel pour chaque obligation
market_prices = [950, 980, 970]

# 2. Fonction pour calculer le rendement approximatif (YTM)
def calculate_ytm(bond, market_price):
    C = bond['face_value'] * bond['coupon_rate']
    F = bond['face_value']
    n = bond['years_to_maturity']
    P = market_price

    # Approximation du YTM
    ytm = (C + (F - P) / n) / ((F + P) / 2)
    return ytm * 100  # en pourcentage

# 3. Calcul des YTM pour toutes les obligations
ytms = [calculate_ytm(bond, price) for bond, price in zip(bonds, market_prices)]
for bond, ytm in zip(bonds, ytms):
    bond['ytm'] = ytm
    print(f"{bond['name']}: YTM = {ytm:.2f}%")

# 4. Comparer les obligations par YTM
bonds_sorted = sorted(bonds, key=lambda x: x['ytm'], reverse=True)
print("\nObligations classées par YTM (décroissant) :")
for bond in bonds_sorted:
    print(f"{bond['name']} - YTM : {bond['ytm']:.2f}%")

# 5. Analyse supplémentaire : prix historiques
historical_prices = np.array([
    [100.0, 200.0, 150.0],  # Day 1 prices
    [101.0, 202.0, 151.0],  # Day 2 prices
    [99.0, 198.0, 149.0]    # Day 3 prices
])

# Moyenne et variation pour chaque obligation
mean_prices = np.mean(historical_prices, axis=0)
price_changes = np.ptp(historical_prices, axis=0)  # peak-to-peak variation

print("\nAnalyse des prix historiques :")
for i, bond in enumerate(bonds):
    print(f"{bond['name']} - Prix moyen : {mean_prices[i]:.2f}, Variation : {price_changes[i]:.2f}")
```

```python

days = np.arange(1, historical_prices.shape[0]+1)

# ---------- GRAPHIQUES ----------

plt.figure(figsize=(12, 5))

# Graphique 1 : YTM des obligations
plt.subplot(1, 2, 1)
names = [bond['name'] for bond in bonds]
plt.bar(names, ytms, color=['skyblue', 'orange', 'green'])
plt.title("Rendement à l'échéance (YTM)")
plt.ylabel("YTM (%)")
plt.ylim(0, max(ytms)+2)

# Graphique 2 : Prix historiques
plt.subplot(1, 2, 2)
for i, bond in enumerate(bonds):
    plt.plot(days, historical_prices[:, i], marker='o', label=bond['name'])
plt.title("Évolution des prix historiques")
plt.xlabel("Jour")
plt.ylabel("Prix")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 5. Mettre en place un outil simple de répartition de l’actif :
#- Stocker les allocations actuelles et cibles dans des baies
#- Calculer les écarts de répartition
#- Déterminer les besoins de rééquilibrage
# Exemple de solution ci-dessus:
# Allocations actuelles et cibles
current_allocations = {
    'Stocks': 0.6,
    'Bonds': 0.3,
    'Cash': 0.1
}
target_allocations = {
    'Stocks': 0.5,
    'Bonds': 0.4,
    'Cash': 0.1
}
# Calcul des écarts
allocation_gaps = {asset: target_allocations[asset] - current_allocations.get(asset, 0)
                   for asset in target_allocations}
print("Allocation Gaps:", allocation_gaps)
# Déterminer les besoins de rééquilibrage
rebalancing_needs = {asset: gap for asset, gap in allocation_gaps.items() if abs(gap) > 0.05}
print("Rebalancing Needs:", rebalancing_needs)
```

```python
# 6. Créez un calculateur de risque de base :
#- Stocker les retours historiques dans des tableaux NumPy
#- Calculer la corrélation entre les actifs
#- Calcul de la volatilité du portefeuille
import numpy as np
# Simulated historical returns for 3 assets over 10 days
historical_returns = np.array([
    [0.01, 0.02, -0.01],  # Day 1 returns
    [0.015, 0.018, -0.005],  # Day 2 returns
    [-0.005, 0.01, 0.002],   # Day 3 returns
    [0.02, -0.01, 0.003],    # Day 4 returns
    [0.01, 0.015, -0.002],   # Day 5 returns
    [0.005, 0.02, 0.001],    # Day 6 returns
    [-0.01, -0.005, 0.004],  # Day 7 returns
    [0.012, 0.017, -0.003],  # Day 8 returns
    [0.008, 0.019, 0.002],   # Day 9 returns
    [0.011, -0.002, -0.001]   # Day 10 returns
])
# Calcul de la matrice de corrélation
correlation_matrix = np.corrcoef(historical_returns, rowvar=False)
print("Correlation Matrix:\n", correlation_matrix)
# Poids du portefeuille
weights = np.array([0.5, 0.3, 0.2])  # 50% Asset 1, 30% Asset 2, 20% Asset 3
# Calcul de la volatilité du portefeuille
portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(np.cov(historical_returns, rowvar=False), weights))) * np.sqrt(252)  # Annualized
print("Portfolio Volatility (Annualized):", portfolio_volatility)
```
