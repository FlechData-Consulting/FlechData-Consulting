## Solution

```python
# 1. Écrivez un programme qui :Prend le cours d’une action en entrée

        # - Si le prix > 100, imprimez « Stock à prix élevé »
        #- Si 50 ≤ prix ≤ 100, imprimez « Stock à prix moyen »
        #- Si le prix < 50, imprimez « Stock à bas prix »
prix_action = float(input("Entrez le cours de l'action : "))
if prix_action > 100:
    print("Stock à prix élevé")
elif 50 <= prix_action <= 100:
    print("Stock à prix moyen")
else:
    print("Stock à bas prix")
```

```python
# 2. Créez un programme qui calcule les intérêts composés sur plusieurs années à l’aide d’une boucle
capital_initial = float(input("Entrez le capital initial : "))  # 1000
taux_interet = float(input("Entrez le taux d'intérêt annuel (en pourcentage) : "))  # 5
annees = int(input("Entrez le nombre d'années : "))  # 10
capital = capital_initial
for annee in range(1, annees + 1):
    capital *= (1 + taux_interet / 100)
    print(f"Année {annee}: Capital = {capital:.2f} euros")
```

```python
#3. Écrivez un programme qui analyse une liste de cours d’actions et :

       # - Trouver les prix les plus élevés et les plus bas
        #- Calcule le prix moyen
        #- Compte le nombre de jours pendant lesquels le prix a été supérieur à la moyenne
cours_actions = [120, 85, 60, 150, 45, 90, 200, 30]  # Exemple de liste de cours d'actions
prix_max = max(cours_actions)
prix_min = min(cours_actions)
prix_moyen = sum(cours_actions) / len(cours_actions)
jours_au_dessus_moyenne = sum(1 for prix in cours_actions if prix > prix_moyen)
print(f"Prix le plus élevé : {prix_max}")
print(f"Prix le plus bas : {prix_min}")
print(f"Prix moyen : {prix_moyen:.2f}")
print(f"Nombre de jours avec un prix supérieur à la moyenne : {jours_au_dessus_moyenne}")
```

```python
#4. Créez un simulateur de trading simple qui :

        #- Passe en revue les prix quotidiens
        #- Utilise les instructions if pour prendre des décisions d’achat/vente
        #- Suit la valeur du portefeuille
prix_quotidiens = [100, 105, 102, 108, 110, 107, 115]  # Exemple de liste de prix quotidiens
capital_initial = 1000  # Capital initial en euros
actions_possedees = 0
capital = capital_initial
for prix in prix_quotidiens:
    if prix < 105 and capital >= prix:  # Acheter si le prix est inférieur à 105 et qu'il y a assez de capital
        actions_possedees += 1
        capital -= prix
        print(f"Acheté une action à {prix}, Actions possédées : {actions_possedees}, Capital restant : {capital:.2f}")
    elif prix > 110 and actions_possedees > 0:  # Vendre si le prix est supérieur à 110 et qu'il y a des actions à vendre
        actions_possedees -= 1
        capital += prix
        print(f"Vendu une action à {prix}, Actions possédées : {actions_possedees}, Capital restant : {capital:.2f}")
print(f"Valeur finale du portefeuille : {capital + actions_possedees * prix_quotidiens[-1]:.2f} euros")
```

```python
#5. Rédiger un programme qui calcule le rendement total d’un investissement avec des cotisations mensuelles sur une période de plusieurs années
capital_initial = float(input("Entrez le capital initial : "))  # 1000
cotisation_mensuelle = float(input("Entrez la cotisation mensuelle : "))  # 100
taux_interet = float(input("Entrez le taux d'intérêt annuel (en pourcentage) : "))  # 5
annees = int(input("Entrez le nombre d'années : "))  # 10
capital = capital_initial
for mois in range(1, annees * 12 + 1):
    capital *= (1 + taux_interet / 100 / 12)  # Appliquer l'intérêt mensuel
    capital += cotisation_mensuelle  # Ajouter la cotisation mensuelle
print(f"Capital après {annees} ans : {capital:.2f} euros")
```

```python
# 6. Créez une calculatrice de remboursement de prêt qui affiche le solde restant après chaque paiement à l’aide d’une boucle

# Calculez les paiements mensuels du prêt en fonction du capital, du taux d’intérêt et des années
# Affichez le solde restant après chaque paiement
capital_emprunte = float(input("Entrez le capital emprunté : "))#200000
taux_annuel = float(input("Entrez le taux d'intérêt annuel (en pourcentage) : "))# 5
duree_annees = int(input("Entrez la durée du prêt (en années) : "))#30
taux_mensuel = taux_annuel / 100 / 12 # conversion du taux annuel en taux mensuel
nombre_paiements = duree_annees * 12 # nombre total de paiements mensuels
paiement_mensuel = (capital_emprunte * taux_mensuel) / (1 - (1 + taux_mensuel) ** -nombre_paiements)
print("Le paiement mensuel du prêt est de : " + str(paiement_mensuel) + " euros.")
for paiement in range(1, nombre_paiements + 1):
    interet = capital_emprunte * taux_mensuel
    principal = paiement_mensuel - interet
    capital_emprunte -= principal
    print(f"Après le paiement {paiement}, le solde restant est de : {capital_emprunte:.2f} euros.")
```
